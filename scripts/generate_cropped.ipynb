```python
import numpy as np
import cv2
from PIL import Image
import os
```

```python
def crop_images(directory, desired_directory, desired_number, ts, to):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".png") and "color" in file:
                filename = os.path.join(root, file)
                dirname = os.path.basename(root)
                name, ext = os.path.splitext(file)
                parts = name.split('_')
                if len(parts) == 2 and dirname == desired_directory and parts[1] == desired_number:
                # if len(parts) == 2 and parts[1] == desired_number:
                    image = cv2.imread(filename)
                    if image is not None:
                        height, width = image.shape[:2]
                        if width >= ts[0] and height >= ts[1]:
                            cropped_image = image[to[1] : ts[1] + to[1], to[0] : ts[0] + to[0]]
                            cropped_filename = os.path.join(root, f"{name}_cropped{ext}")
                            cv2.imwrite(cropped_filename, cropped_image)
                            print(f"Cropped image saved: {cropped_filename}")
```

```python
directory = 'results_png/mip360'  # Replace with the desired directory path
desired_directory = 'csn'  # Replace with the desired directory name
desired_number = '002'  # Replace with the desired number
target_size = (400, 300)  # Replace with the desired target size (width, height)
target_offset = (980, 350)

crop_images(directory, desired_directory, desired_number, target_size, target_offset)
```

```python
def convert_png_to_jpg(directory, output_directory):
    for root, dirs, files in os.walk(directory):
        print(root)
        for file in files:
            if file.endswith(".png"):
                source_path = os.path.join(root, file)
                rel_path = os.path.relpath(source_path, directory)
                output_path = os.path.join(output_directory, rel_path)
                os.makedirs(os.path.dirname(output_path), exist_ok=True)
                output_file = os.path.splitext(output_path)[0] + ".jpg"

                output_file = output_file.replace('cropped_000', 'cropped')
                output_file = output_file.replace('cropped_001', 'cropped')
                output_file = output_file.replace('cropped_002', 'cropped')
                output_file = output_file.replace('cropped_050', 'cropped')
                output_file = output_file.replace('cropped_100', 'cropped')

                image = Image.open(source_path)
                rgb_image = image.convert("RGB")
                rgb_image.save(output_file, "JPEG")
                print(f"Converted: {output_file}")
```

```python
source_directory = 'results_png/mip360/'  # Replace with the source directory path
output_directory = 'results_jpg/mip360/'  # Replace with the output directory path

convert_png_to_jpg(source_directory, output_directory)
```

```python
def fix_images(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".png"):
                filename = os.path.join(root, file)
                image = cv2.imread(filename)
                if image is not None:
                    fixed_image = image
                    cv2.imwrite(filename, fixed_image)
                    print(f"Fixed image saved: {filename}")
```
